```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# import dataset
df_minor = pd.read_csv('road_issues_minor.csv',index_col=0)

lower_case_cols = [col_name.lower() for col_name in df_minor.columns]
col_names = ['_'.join(col_name.split(' ')) for col_name in lower_case_cols]
df_minor.columns = col_names
```

```python
df_minor.info()
```

```python
df_minor.date_initiated = pd.PeriodIndex(df_minor.date_initiated,freq='d')
df_minor.date_time_closed = pd.PeriodIndex(df_minor.date_time_closed,freq='d')
```
